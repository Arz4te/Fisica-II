# Clase1 - Introducción

La **mecánica clásica** o **mecánica newtoniana**  es la rama de la física que estudia la  *dinámica* (movimiento o reposo) e interacción de los cuerpos macroscópicos, y su evolución en el tiempo, bajo la acción de fuerzas

Sus métodos fueron desarrolados por Isaac Newton y posteriormente nutridos y establecidos por Leibniz, Lagrange, Euler, y contemporáneos, en el siglo XVII.


Newton.jpg


Bibliografía de [Isaac Newton](https://www.youtube.com/watch?v=uw2WhlxDdRQ)




Newton desarrolla 3 leyes de movimiento de la mecánica clásica:
1. Todo cuerpo continúa en su estado de reposo o movimiento uniforme a menos 
que se ejerza una fuerza sobre dicho cuerpo.
2. $\vec{F} = m\vec{a}$
3. A toda acción corresponde una reacción con la misma magnitud pero en sentido apuesto


Es necesario definir conceptos como materia y fuerza para entender mejor la 2da ley de Newton. La **materia** se define  como todo aquello que ocupa un espacio en el universo, es decir, lo que posee masa y volumen.  La **fuerza** es una cantidad vectorial que se mide en newtons ($N = kg \cdot m/s^2$) y se puede intepretar como la tasa de cambio temporal del momento lineal de una masa o partícula o como la capacidad de deformar un material.
